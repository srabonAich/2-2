## Introduction

A **2's Complement Adder-Subtractor** is a digital circuit that can perform **both addition and subtraction** of binary numbers using the **2's complement method**. The same hardware (mainly an adder) is reused for both operations, making it efficient in terms of resources.

---

## Concept of 2's Complement

- The 2's complement of a binary number is obtained by:
  1. Inverting all bits (1's complement)
  2. Adding 1 to the least significant bit

- To subtract B from A using 2’s complement:
- A - B = A + (2’s complement of B)


---

## Working Principle

The circuit consists of:
1. **XOR gates**: Used to conditionally invert the bits of B.
2. **Full Adders**: Perform the addition of A and modified B.
3. **Control Line (SUB)**: Determines the operation.
 - If SUB = 0 → Perform A + B
 - If SUB = 1 → Perform A - B = A + (~B + 1)

---

## Bit-Level Operation

| SUB (Control Bit) | Operation |
|-------------------|-----------|
| 0                 | A + B     |
| 1                 | A + 2's complement of B = A - B |

- Each bit of B passes through an XOR gate:
  - B[i] XOR SUB → inverts B[i] if SUB = 1
- SUB is also connected to the **carry-in** of the LSB full adder (to add the +1 in 2's complement).

---

## Full Adder Logic (per bit)

Let A[i], B[i] be the input bits and SUB be the control signal:

- Modified B[i] = B[i] ⊕ SUB
- Carry-in (initial) = SUB

- Sum[i] = A[i] ⊕ Modified B[i] ⊕ Carry[i]
- Carry-out = Majority(A[i], Modified B[i], Carry[i])

---

## Example

### A = `0101` (5), B = `0011` (3)

- **Addition**: SUB = 0  
  - Result = `0101` + `0011` = `1000` (8)

- **Subtraction**: SUB = 1  
  - B = `0011` → Invert = `1100`, Add 1 = `1101`
  - Result = `0101` + `1101` = `10010` → Truncate to 4 bits = `0010` (2)

---

## Advantages

- Uses a **single adder** for both addition and subtraction.
- Efficient hardware design with minimal components.

---

## Disadvantages

- Limited to 2's complement representation.
- For overflow detection, extra logic is needed.

---

## Applications

- ALUs (Arithmetic Logic Units)
- Digital signal processors
- CPUs and embedded processors

---

## Conclusion

A 2's Complement Adder-Subtractor circuit effectively performs both binary addition and subtraction using the 2's complement method. It’s a core element of many arithmetic logic circuits due to its simplicity and efficiency.
