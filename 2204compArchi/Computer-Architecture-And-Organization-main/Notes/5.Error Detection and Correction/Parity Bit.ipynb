## üîç Introduction

**Parity bit error detection** is a simple method used in digital communication systems to detect errors in transmitted data. It is one of the most basic forms of error detection.

## üéØ Objective

To detect if an error has occurred during the transmission of binary data using a single bit called the **parity bit**.

---

## üß† Basic Concept

A **parity bit** is an extra bit added to a string of binary data. It helps the receiver determine whether the number of bits with the value `1` is **even** or **odd**.

There are two types of parity:

1. **Even Parity**: The total number of 1s in the data (including the parity bit) is even.
2. **Odd Parity**: The total number of 1s in the data (including the parity bit) is odd.

---

## üì¶ How it Works

### Example:

Consider the 7-bit data: `1011001`

- **Even Parity**: There are 4 ones. It‚Äôs already even ‚Üí Add parity bit = `0`
  - Transmitted: `10110010`
  
- **Odd Parity**: 4 ones ‚Üí Add parity bit = `1` (to make it odd)
  - Transmitted: `10110011`

---

## ‚úÖ Error Detection

When the data is received:

1. The receiver counts the number of `1s` (including the parity bit).
2. Based on the parity type, it checks:
   - For **even parity**: Is the number of 1s even?
   - For **odd parity**: Is the number of 1s odd?
3. If the parity does not match, **an error is detected**.

‚ö†Ô∏è **Limitation**: Parity bits can **only detect single-bit errors**. They cannot correct errors or detect all multi-bit errors.

## üîÅ Pros and Cons

### ‚úÖ Advantages
- Simple to implement
- Requires only 1 additional bit
- Effective for single-bit error detection

### ‚ùå Limitations
- Cannot detect all multi-bit errors
- Cannot correct any errors

---

## üß™ Use Cases

- UART (Universal Asynchronous Receiver-Transmitter)
- Simple sensor-to-controller communications
- Legacy systems with low reliability requirements

---

## üß© Summary

- Parity bits help detect **single-bit errors** using **even or odd parity** rules.
- They are **not sufficient** for complex error environments.
- Still used in many **low-complexity systems** for basic data integrity checking.
